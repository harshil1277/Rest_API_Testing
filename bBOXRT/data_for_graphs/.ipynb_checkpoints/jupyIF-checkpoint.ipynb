```python
# counts num pass per operation
pathName = "/Users/thedemean/Documents/blackbox_git_dei_inv/bBOXRT/data_for_graphs/tpcc/case5_With_Distance_in_content_Response/"
fileNameif = "data_if_TPC-C Rest API_0.txt"
df = pd.read_csv(pathName + fileNameif, sep="\t", header=None)
population_size = len(df.iloc[0,0:])-1
df_finalStatusCode = []
db_finalStatusCode = []
num_of_experiences = 30
# number of operations 
num_rows_operations = int(len(df[0]) / 3)
total_max_pass = num_of_experiences*population_size

df_mean_fitness_per_operation = []
db_mean_fitness_per_operation = []

df_mean_statuscode_per_operation = []
db_mean_statuscode_per_operation = []

df_mean_distance_per_operation = []
db_mean_distance_per_operation = []


df_total_pass_per_operation = []
db_total_pass_per_operation = []
db_total_pass_per_operation_per_run = []
df_total_pass_per_operation_per_run = []
for i in range(num_rows_operations):
    df_mean_fitness_per_operation.append(0)
    db_mean_fitness_per_operation.append(0)
    df_mean_statuscode_per_operation.append(0)
    db_mean_statuscode_per_operation.append(0)
    df_mean_distance_per_operation.append(0)
    db_mean_distance_per_operation.append(0)
    
    df_total_pass_per_operation.append(0)
    db_total_pass_per_operation.append(0)
    db_total_pass_per_operation_per_run.append([])
    df_total_pass_per_operation_per_run.append([])
    
    for k in range(num_of_experiences):
        db_total_pass_per_operation_per_run[i].append(0)
        df_total_pass_per_operation_per_run[i].append(0)

total_pass_if = 0
total_pass_rs = 0
mean_fitness_if = 0.0
mean_fitness_bs = 0.0
                #30
for i in range (num_of_experiences):
    fileNameif = "data_if_TPC-C Rest API_" + str(i) + ".txt"
    fileNamebs = "data_bs_TPC-C Rest API_" + str(i) + ".txt"

    df = pd.read_csv(pathName + fileNameif, sep="\t", header=None)
    db = pd.read_csv(pathName + fileNamebs, sep="\t", header=None)
    
    index = num_rows_operations # = 4 to tpc-c
    index_distance = num_rows_operations*2
    index_fitness = 0
    total_pass_in_run_if = 0
    total_pass_in_run_rs = 0
    for j in range(num_rows_operations):
        for k in range (len(df.iloc[index,1:])):
            df_mean_fitness_per_operation[j] += df.iloc[index_fitness,k]
            db_mean_fitness_per_operation[j] += db.iloc[index_fitness,k]
            
            df_mean_statuscode_per_operation[j] += df.iloc[index,k]
            db_mean_statuscode_per_operation[j] += db.iloc[index,k]
            
            df_mean_distance_per_operation[j] += df.iloc[index_distance,k]
            db_mean_distance_per_operation[j] += db.iloc[index_distance,k]
            
            # note that 0.8 is when status code weight is equal to 0.8, if 0.7 then this to change
            if  df.iloc[index,k] >= 0.5:
                df_total_pass_per_operation[j] = df_total_pass_per_operation[j] + 1
                total_pass_in_run_if = total_pass_in_run_if + 1
            if  db.iloc[index,k] >= 0.5:
                db_total_pass_per_operation[j] = db_total_pass_per_operation[j] + 1
                total_pass_in_run_rs = total_pass_in_run_rs + 1
        index += 1
        index_fitness += 1
        index_distance += 1
        
        # change db_total_pass_per_operation or df_total_pass_per_operation for random or informative
        df_total_pass_per_operation_per_run[j][i] = df_total_pass_per_operation[j]
        db_total_pass_per_operation_per_run[j][i] = db_total_pass_per_operation[j]
    #print(df_total_pass_per_operation)
    for j in range(num_rows_operations):
        df_total_pass_per_operation[j]=0
        db_total_pass_per_operation[j] = 0
    total_pass_if = total_pass_if + total_pass_in_run_if
    total_pass_rs = total_pass_rs + total_pass_in_run_rs

    
        
```

```python
# print from previous cell
print("IF fitness mean per operation: ")
for j in range(num_rows_operations):
    df_mean_fitness_per_operation[j] = df_mean_fitness_per_operation[j]/(num_of_experiences*population_size)
print(df_mean_fitness_per_operation)

print("IF statuscode mean per operation: ")
for j in range(num_rows_operations):
    df_mean_statuscode_per_operation[j] = df_mean_statuscode_per_operation[j]/(num_of_experiences*population_size)
print(df_mean_statuscode_per_operation)

print("IF distance mean per operation: ")
for j in range(num_rows_operations):
    df_mean_distance_per_operation[j] = df_mean_distance_per_operation[j]/(num_of_experiences*population_size)
print(df_mean_distance_per_operation)
    
    
    
print("BS fitness mean per operation: ")
for j in range(num_rows_operations):
    db_mean_fitness_per_operation[j] = db_mean_fitness_per_operation[j]/(num_of_experiences*population_size)
print(db_mean_fitness_per_operation)

print("BS statuscode mean per operation: ")
for j in range(num_rows_operations):
    db_mean_statuscode_per_operation[j] = db_mean_statuscode_per_operation[j]/(num_of_experiences*population_size)
print(db_mean_statuscode_per_operation)

print("BS distance mean per operation: ")
for j in range(num_rows_operations):
    db_mean_distance_per_operation[j] = db_mean_distance_per_operation[j]/(num_of_experiences*population_size)
print(db_mean_distance_per_operation)
```

```python
import numpy as np
from scipy import stats
from scipy.stats import mannwhitneyu

alpha = 0.05
n_sided = 1 # 2-sided test

z_crit = stats.norm.ppf(1-alpha/n_sided)
print("z_crit border:" , z_crit) # 1.959963984540054
# critical to alpha
#alpha = stats.norm.sf(z_crit) * n_sided
#print(alpha) # 0.05

for j in range(num_rows_operations):
    x = np.array(df_total_pass_per_operation_per_run[j])
    y = np.array(db_total_pass_per_operation_per_run[j])
    print("---------------------------------------")
    
    if np.array_equal(x,y):
        print("Sets are equal")
        continue
    res = mannwhitneyu(x, y, use_continuity=True, alternative='greater')
    
    z_crit = stats.norm.ppf(1-res.pvalue/n_sided)
    print("pvalue: ", res.pvalue)
    print("z_crit: ", z_crit)
print("---------------------------------------")
```
